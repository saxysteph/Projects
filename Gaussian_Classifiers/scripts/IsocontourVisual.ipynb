```python
import sys
if sys.version_info[0] < 3:
	raise Exception("Python 3 not detected.")
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from scipy import io
import pandas as pd
from scipy.stats import multivariate_normal
```

1)

```python
mean = [1,1]
cov = [[1, 0], [0, 2]]

x, y = np.meshgrid(np.linspace(-3, 5, 100), np.linspace(-3, 5, 100))
pos = np.dstack((x, y))

pdfVal = multivariate_normal(mean, cov).pdf(pos)

plt.figure(figsize = (8, 6))
plt.contour(x,y, pdfVal, levels = np.linspace(0, pdfVal.max(), 25))
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Isocontours - Problem 1')
plt.colorbar(label = 'Isovalue')
plt.show
```

```python
mean1 = [0,2]
mean2 = [2,0]
cov = [[2, 1], [1, 1]]

x, y = np.meshgrid(np.linspace(-5, 5, 100), np.linspace(-3, 7, 100))
pos = np.dstack((x, y))

pdfVal1 = multivariate_normal(mean1, cov).pdf(pos)
pdfVal2 = multivariate_normal(mean2, cov).pdf(pos)

plt.figure(figsize = (8, 6))
plt.contour(x,y, pdfVal1 - pdfVal2, levels = np.arange(-0.10, 0.10, 0.01))
#plt.contour(x,y, pdfVal1 - pdfVal2, levels = np.linspace(-0.1, pdfVal2.max(), 25))
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Isocontours - Problem 3')
plt.colorbar(label = 'Isovalue')
plt.show
```
